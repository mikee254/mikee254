```python
# @title Car make vs Colour

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Colour'].value_counts()
    for x_label, grp in cardata.groupby('Car make')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Car make')
_ = plt.ylabel('Colour')
```
