# matplotlib example workflow

```python
# 0 import matplotlib and get ready for plotting
%matplotlib inline
import matplotlib.pyplot as plt

#1 prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2 setup plot
fig,ax=plt.subplots(figsize=(5,5))

#3 plot data
ax.plot(x,y)
# customize plot
ax.set(title="simple plot",
       xlabel="x-axis",
       ylabel="y-axis")
# save & show (you save the whole figure)
```

## making figures with numpy arrays

we want
* line plot
* scatter plot
* bar plot
* histogram
* subplots

```python
import numpy as np
# create some data
x=np.linspace(0,10,100)
x[:10]
```

```python
# plot the data and create a line plot
fir, ax=plt.subplots()
ax.plot(x,x**2); # to power 2
```

```python
# use same data to make a scatter
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# another scatter
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# another scatter
fig,ax=plt.subplots()
ax.scatter(x,np.cos(x));
```

```python
# make a plot from dictionary
nutbutterprices={"Almond":300,"Peanut":250,"Cashew":500}
fig,ax=plt.subplots()
ax.bar(nutbutterprices.keys(),nutbutterprices.values());
ax.set(title="MIKE's Nut Butter Store",
      ylabel="Price (ksh)");
```

```python
fig, ax=plt.subplots()
ax.barh(list(nutbutterprices.keys()),list(nutbutterprices.values()));
```

```python
# make some data for histogram and plot it
x=np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

# two options for subplots

```python
# subplot option 1
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
# plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(100),np.random.random(100));
ax3.bar(nutbutterprices.keys(),nutbutterprices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig,ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))
# plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(100),np.random.random(100));
ax[1,0].bar(nutbutterprices.keys(),nutbutterprices.values());
ax[1,1].hist(np.random.randn(1000));
```
