# two options for subplots

```python
# subplot option 1
fig, ((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
# plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(100),np.random.random(100));
ax3.bar(nutbutterprices.keys(),nutbutterprices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig,ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))
# plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(100),np.random.random(100));
ax[1,0].bar(nutbutterprices.keys(),nutbutterprices.values());
ax[1,1].hist(np.random.randn(1000));
```

### which one you use?( pyplot vs matplotlib oo method)
*when plotting something quickly, okay to use pyplot method
*when plotting something more advanced, use the OO method


```python
heartdisease.head()
```

```python
over50=heartdisease[heartdisease["age"]>50]
over50.head()
```

```python
over50.target.values
```

```python
over50.plot(kind="scatter", # pyplot method
            x="age",
            y="chol",
            c="target");
```

```python
# oo method (object oriented)
fix, ax=plt.subplots(figsize=(10,6))
over50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
```

```python
## oo method from scratch
fig, ax=plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(x=over50["age"],
                   y=over50["chol"],
                   c=over50["target"])
# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over50["chol"].mean(),
           linestyle="--");
```

```python
over50.head()
```

```python
# subplots of chol, age, thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex= True)
# add data to ax0
scatter=ax0.scatter(x=over50["age"],
                   y=over50["chol"],
                   c=over50["target"])
# customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",

       ylabel="Cholesterol");
# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= "Target") # removed extra space here
# add a meanline
ax0.axhline (y=over50["chol"].mean(),
               linestyle="--");
# add data to ax1
scatter=ax1.scatter(x=over50["age"],
                    y=over50["thalach"],
                    c=over50["target"])
# customize ax1
ax1.set(title="Heart Diesease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate");
# add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");
# add a mealine
ax1.axhline(y=over50["thalach"].mean(),
            linestyle="--");
# add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```

```python
# set style
plt.style.use('seaborn-whitegrid'); # Use 'seaborn-whitegrid' instead of 'seabornwhitegrid'
## oo method from scratch
fig, ax=plt.subplots(figsize=(10,6));

# plot the data
scatter=ax.scatter(x=over50["age"],
                   y=over50["chol"],
                   c=over50["target"],
                   cmap="summer"); # this changes the colour scheme
# customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over50["chol"].mean(),
           linestyle="--");
```

This plot shows information about heart disease

```python
# subplots of chol, age, thalach
fig, (ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,10),
                           sharex= True)
# add data to ax0
scatter=ax0.scatter(x=over50["age"],
                   y=over50["chol"],
                   c=over50["target"],
                    cmap="winter");
# customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",

       ylabel="Cholesterol");
 # change the x axis limit (removes the double line)
ax0.set_xlim([50, 80])
# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title= "Target") # removed extra space here
# add a meanline
ax0.axhline (y=over50["chol"].mean(),
               linestyle="--");
# add data to ax1
scatter=ax1.scatter(x=over50["age"],
                    y=over50["thalach"],
                    c=over50["target"],
                    cmap="summer");
# customize ax1
ax1.set(title="Heart Diesease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate");
ax0.set_xlim([50, 80])
ax1.set_ylim([60, 200])
# add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target");
# add a mealine
ax1.axhline(y=over50["thalach"].mean(),
            linestyle="--");
# add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold")
```
