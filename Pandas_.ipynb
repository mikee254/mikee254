```python
import pandas as pd

# Load the data
carsales = pd.DataFrame({
    "Make": ["Toyota", "Honda", "Ford", "Chevy", "Black"],
    "Colour": ["Red", "Blue", "White", "Silver", "Black"],
    "Odometer (KM)": [100000, 50000, 30000, 20000, 10000],
    "Doors": [4, 4, 4, 4, 2]
})

# Group by Make and calculate the mean of numeric columns only
numeric_columns = ["Odometer (KM)", "Doors"]
grouped_data = carsales.groupby("Make")[numeric_columns].mean(numeric_only=True)

# Print the results
print(grouped_data)
```
